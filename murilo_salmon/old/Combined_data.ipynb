```python
import pandas as pd
Combined_date = pd.read_csv(r'C:\Users\MuriloFarias\Desktop\NNS-JULIA\PredictSalmonRuns\data\Combined_FeatureSet_For_Model.csv')

```

```python
# Filter for River == 'Bonneville Lock & Dam'
filtered_df = Combined_date[Combined_date['River'] == 'Bonneville Lock & Dam']

# Get unique Years and sort
years = filtered_df[['Year']].drop_duplicates().sort_values(by='Year').reset_index(drop=True)

# Show result
years
```

```python

# Turn off scientific notation for floats
pd.set_option('display.float_format', '{:,.0f}'.format)

# Group and sort
grouped = Combined_date.groupby(['System', 'River'])['Total_Returns'].sum().reset_index()
grouped = grouped.sort_values(by=['System', 'River', 'Total_Returns'])

# Show result
grouped
```

```python
ageclass_cols = [
    'AgeClass_0.1', 'AgeClass_0.2', 'AgeClass_0.3', 'AgeClass_0.4', 'AgeClass_0.5',
    'AgeClass_1.1', 'AgeClass_1.2', 'AgeClass_1.3', 'AgeClass_1.4', 'AgeClass_1.5',
    'AgeClass_2.1', 'AgeClass_2.2', 'AgeClass_2.3', 'AgeClass_2.4',
    'AgeClass_3.1', 'AgeClass_3.2', 'AgeClass_3.3', 'AgeClass_3.4'
]

# Group and sum
grouped = Combined_date.groupby(['System', 'River'])[ageclass_cols].sum()

# Sum across AgeClass columns to get Total_sum
grouped['Total_sum'] = grouped.sum(axis=1)

# Keep only System, River, and Total_sum
grouped = grouped[['Total_sum']].reset_index()

# Optional: sort
grouped = grouped.sort_values(by=['System', 'River'])

# Show result
grouped


# Group and sum
grouped = Combined_date.groupby(['System', 'River'])[ageclass_cols].sum()

# Sum across AgeClass columns to get Total_sum
grouped['Total_sum'] = grouped.sum(axis=1)

# Keep only System, River, and Total_sum
grouped = grouped[['Total_sum']].reset_index()

# Optional: sort
grouped = grouped.sort_values(by=['System', 'River'])

# Show result
grouped
```
