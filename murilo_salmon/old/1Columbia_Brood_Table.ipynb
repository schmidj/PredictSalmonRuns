```python
# --- STEP 1: Select relevant age class columns ---
age_class_columns = [col for col in columbia_brood.columns if col.startswith('AgeClass_')]
cols_to_keep = ['BroodYear', 'Total_Recruits'] + age_class_columns
clean_df_columbia_brood = columbia_brood[cols_to_keep].copy()

# --- STEP 2: Melt to long format ---
clean_df_columbia_brood_long = clean_df_columbia_brood.melt(
    id_vars=['BroodYear', 'Total_Recruits'],
    value_vars=age_class_columns,
    var_name='AgeClass',
    value_name='RecruitCount'
)
```

```python
# STEP 1: Keep only relevant columns
age_class_columns = [col for col in clean_df_columbia_brood.columns if col.startswith('AgeClass_')]
cols_to_keep = ['BroodYear', 'Total_Recruits'] + age_class_columns
clean_df_columbia_return = clean_df_columbia_brood[cols_to_keep].copy()


# STEP 3: Remove age class columns where all values are 0
age_class_columns = [col for col in age_class_columns if clean_df_columbia_return[col].sum() > 0]

# STEP 4: Melt to long format
clean_df_columbia_return_long = clean_df_columbia_return.melt(
    id_vars=['BroodYear'],
    value_vars=age_class_columns,
    var_name='AgeClass',
    value_name='Total_age'
)
```
