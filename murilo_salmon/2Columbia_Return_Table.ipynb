```python
# STEP 1: Keep only relevant columns
age_class_columns = [col for col in columbia_return_df.columns if col.startswith('AgeClass_')]
cols_to_keep = ['ReturnYear', 'Total_Returns'] + age_class_columns
clean_df_columbia_return = columbia_return_df[cols_to_keep].copy()

# STEP 2: Compute Unclassified before dropping anything
clean_df_columbia_return['Unclassified'] = (
    clean_df_columbia_return['Total_Returns'] -
    clean_df_columbia_return[age_class_columns].fillna(0).sum(axis=1)
).clip(lower=0)  # force non-negative

# STEP 3: Remove age class columns where all values are 0
age_class_columns = [col for col in age_class_columns if clean_df_columbia_return[col].sum() > 0]

# STEP 4: Melt to long format
clean_df_columbia_return_long = clean_df_columbia_return.melt(
    id_vars=['ReturnYear'],
    value_vars=age_class_columns,
    var_name='AgeClass',
    value_name='ReturnCount'
)
```

```python
#AgeClass_Unclassified	

# 5. Define age class columns from the original dataset
columns_to_sum = [
    'AgeClass_0.1', 'AgeClass_0.2', 'AgeClass_0.3', 'AgeClass_0.4', 'AgeClass_0.5',
    'AgeClass_1.1', 'AgeClass_1.2', 'AgeClass_1.3', 'AgeClass_1.4', 'AgeClass_1.5',
    'AgeClass_2.1', 'AgeClass_2.2', 'AgeClass_2.3', 'AgeClass_2.4',
    'AgeClass_3.1', 'AgeClass_3.2', 'AgeClass_3.3', 'AgeClass_3.4'
]

# 6. Get Total_Returns per year from original data
total_returns_by_year = columbia_return_df.set_index('ReturnYear')['Total_Returns']

# 7. Calculate sum of actual age classes used (excluding 'Unclassified')
ageclass_sum_by_year = clean_df_columbia_return_long[
    clean_df_columbia_return_long['AgeClass'] != 'AgeClass_Unclassified'
].groupby('ReturnYear')['ReturnCount'].sum()

# 8. Compute unclassified = Total - sum of actual age classes used
unclassified_by_year = (total_returns_by_year - ageclass_sum_by_year).clip(lower=0)

# 9. Filter only those years where unclassified > 0
unclassified_df = unclassified_by_year[unclassified_by_year > 0].reset_index()
unclassified_df['AgeClass'] = 'AgeClass_Unclassified'
unclassified_df.rename(columns={0: 'ReturnCount', 'Total_Returns': 'ReturnCount'}, inplace=True)
unclassified_df['ReturnCount'] = unclassified_by_year[unclassified_by_year > 0].values

# 10. Reorder columns
unclassified_df = unclassified_df[['ReturnYear', 'AgeClass', 'ReturnCount']]

# 11. Append to clean long-format DataFrame
clean_df_columbia_return_long = pd.concat([clean_df_columbia_return_long, unclassified_df], ignore_index=True)
```

```python
# STEP 12: Extract numeric part from AgeClass, handling 'Unclassified'
def extract_age_value(age_class):
    try:
        return float(age_class.replace('AgeClass_', ''))
    except ValueError:
        return None

clean_df_columbia_return_long['Age'] = clean_df_columbia_return_long['AgeClass'].apply(extract_age_value)

# Calculate TotalAge as integer, or None
def calculate_total_age(age_float):
    if pd.isna(age_float):
        return None
    parts = str(age_float).split('.')
    return int(parts[0]) + int(parts[1]) if len(parts) == 2 else int(parts[0])

clean_df_columbia_return_long['TotalAge'] = clean_df_columbia_return_long['Age'].apply(calculate_total_age)

# STEP 12: Calculate BroodYear only when TotalAge exists
clean_df_columbia_return_long['BroodYear'] = clean_df_columbia_return_long.apply(
    lambda row: row['ReturnYear'] - row['TotalAge'] if row['TotalAge'] is not None else None,
    axis=1
)

# Cast both as int (not float)
# Convert columns to string and replace missing values with '-'
clean_df_columbia_return_long['Age'] = clean_df_columbia_return_long['Age'].apply(lambda x: '-' if pd.isna(x) else str(round(x, 1)))
clean_df_columbia_return_long['TotalAge'] = clean_df_columbia_return_long['TotalAge'].apply(lambda x: '-' if pd.isna(x) else str(int(x)))
clean_df_columbia_return_long['BroodYear'] = clean_df_columbia_return_long['BroodYear'].apply(lambda x: '-' if pd.isna(x) else str(int(x)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure TotalAge is numeric
clean_df_columbia_return_long['TotalAge'] = clean_df_columbia_return_long['TotalAge'].astype(float)

# Create the base boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=clean_df_columbia_return_long, x='TotalAge', y='ReturnCount', showfliers=False)

# Add all points as scatter (with some jitter)
sns.stripplot(data=clean_df_columbia_return_long, x='TotalAge', y='ReturnCount',
              color='red', jitter=True, alpha=0.5)

# Calculate IQR for each TotalAge to identify outliers manually
grouped = clean_df_columbia_return_long.groupby('TotalAge')

for age, group in grouped:
    q1 = group['ReturnCount'].quantile(0.25)
    q3 = group['ReturnCount'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = group[(group['ReturnCount'] < lower_bound) | (group['ReturnCount'] > upper_bound)]

    for _, row in outliers.iterrows():
        plt.text(x=float(row['TotalAge']) - 0.2,
                 y=row['ReturnCount'],
                 s=str(int(row['ReturnYear'])),
                 fontsize=8, color='black')

# Finalize plot
plt.title('Distribution of ReturnCount by TotalAge (Labeled Outliers)')
plt.xlabel('Total Age')
plt.ylabel('Return Count')
plt.tight_layout()
plt.show()
```
