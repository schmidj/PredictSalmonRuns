One-Hot Encode River Column

```python
# One-hot encode and convert True/False to 1/0
data_set_bristol_encoded = pd.get_dummies(data_set_bristol, columns=['River'], prefix='River')

```

```python
river_cols = [col for col in data_set_bristol_encoded.columns if col.startswith('River_')]
data_set_bristol_encoded[river_cols] = data_set_bristol_encoded[river_cols].astype(int)
```

Checking if the main values still ok after one hot encode

```python
#original data data_set_bristal
total_return_bristol = data_set_bristol['Total_Returns'].sum()
total_next_bristol = data_set_bristol['Total_Returns_NextYear'].sum()

# after one hot encoding data_set_bristol_encoded
total_return_encoded = data_set_bristol_encoded['Total_Returns'].sum()
total_next_encoded = data_set_bristol_encoded['Total_Returns_NextYear'].sum()

# Differences
diff_return = total_return_encoded - total_return_bristol
diff_next = total_next_encoded - total_next_bristol

print("  Difference in Total_Return:", diff_return)
print("  Difference in Total_Returns_NextYear:", diff_next)
```

Correlation matrix for all tables (high correlation here were removed in Julia Pipe Line) Im just checking again

```python
# Generate correlation matrix
correlation_matrix = data_set_bristol_encoded.corr(numeric_only=True)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f",
            mask=(abs(correlation_matrix) < 0.5))  # Only show >0.9 or <-0.9
plt.title("Correlation Heatmap (>|0.9|) - Highlighted Only")
plt.tight_layout()
plt.show()

# Filter correlation matrix for high correlations (excluding self-correlation)
high_corr_mask = ((correlation_matrix > 0.5) | (correlation_matrix < -0.5)) & (correlation_matrix != 1.0)

# Get column names where at least one high correlation exists
high_corr_columns = high_corr_mask.any(axis=0)
columns_with_high_corr = high_corr_columns[high_corr_columns].index.tolist()

print("Columns with correlation > 0.5 or < -0.5:")
for i in columns_with_high_corr:
    print(f"- {i}")
```

Now this correlation here is between our label (Total_Returns_NextYear ) vs our features ()
the idea its choose the best features

```python
# 1. Calculate correlation matrix
corr_matrix = data_set_bristol_encoded.corr(numeric_only=True)
# 2. Get correlation of all features with the target
target_corr = corr_matrix['Total_Returns_NextYear'].drop('Total_Returns_NextYear')

# 3. Sort by absolute correlation (descending) and print
target_corr_sorted = target_corr.reindex(target_corr.abs().sort_values(ascending=False).index)

print("Correlation with Total_Returns_NextYear (sorted):")
print(target_corr_sorted)
```
